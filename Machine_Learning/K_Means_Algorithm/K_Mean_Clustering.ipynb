```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#blob - a collection of data in binary form, typically a multimedia item, stored as a single entity in a database management system.
#make_blobs = generates isotropic Gaussian blobs for clustering
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
#in real data we can't compare due to we dont have a labels, its unsupervised learning
#but in this example where we created data ourselfs we can compare with data[1]
#KMeans used to try to find labels in data 
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
